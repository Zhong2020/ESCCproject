```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
loh=pd.read_table("data/Fig5BandC.txt")
```

## Figure 5B

```python
ax=loh['region'].value_counts().value_counts().plot(kind="bar",color=sns.color_palette())
ax.set_ylabel("Sample Count",size="large")
ax.set_xlabel("Number of Lost Allele",size="large")
for item in ax.get_xticklabels():
    item.set_rotation(360)
    
# create a list to collect the plt.patches data
totals = []

# find the values and append to list
for i in ax.patches:
    totals.append(i.get_height())

# set individual bar lables using above list
total = 103

# set individual bar lables using above list
for i in ax.patches:
    # get_x pulls left or right; get_height pushes up or down
    ax.text(i.get_x()+.05, i.get_height()-1.5, \
            str(round((i.get_height()/total)*100, 2))+'%', fontsize=12,
                color='white')
ax.set_title("%0.2f%% samples  with HLALOH"%(54./103*100))
plt.tight_layout()
```

## Figure 5C

```python
dat=loh['LossAllele'].str.upper().str.replace("HLA_","").str.replace("A_","A*").str.replace("B_","B*").str.replace("C_","C*").str.replace("_",":").str[:7].value_counts()

ax=dat.plot(kind="bar",title="HLALOH Count",color=sns.color_palette())
for item in ax.get_xticklabels():
    item.set(fontsize=8)
plt.ylabel("Count")
plt.tight_layout()
```
