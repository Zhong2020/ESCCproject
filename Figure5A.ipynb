## read in the data in sheet2 of "gene_set_NMF_4groups.xlsx"

```r
library(ggplot2)
```

```r
dat=data.frame(Geneset=rep(c("Stemness_up","XCL1_up","Immunogenic_up","Metabolic_up","Differentiated_up"),times=2),
value=c(1.36,0.87,-1.92,-3.19,-3.86,2.49,1.47,0.94,2.71,-3.47),
FDR=c(0.05,0.75,6.41E-04,0,0,0,0.014,0.6,0,0),
Sig=c("*","","***","***","***","***","*","","***","***"),
Category=rep(c("Mut vs Wt","High vs Low"),each=5)
)
```

```r
options(repr.plot.width=7,repr.plot.height=5)
ggplot(dat,aes(x=Geneset,y=value,fill=Category))+
    geom_bar(stat="identity",position="dodge")+ 
annotate("text",size=8,color="blue",vjust=0,
         y=c(0.5,0.5,-0.5,0.5,-0.5,0.5,-0.5,-0.5,-0.5),
         x=c(0.5,1,1.5,2,2.5,3,3.5,4,4.5),
    label=c("***","***","","***","***","***","***","*","*"),)+
    labs(x="",y="NES")+
    coord_flip()+
theme_classic()+
theme(axis.text=element_text(size=22),axis.ticks.y = element_blank(),
    axis.line.y = element_blank(),
    axis.title=element_text(size=20),
    axis.text.y=element_text(size=22,angle=0),
    legend.title=element_blank(),
    legend.text=element_text(size=20),
    legend.position="top")
```
